```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

with open("space.txt", "r") as f:
    space_data = [line.strip().split(",")[1:] for line in f.readlines()]

with open("sports.txt", "r") as f:
    sports_data = [line.strip().split(",")[1:] for line in f.readlines()]

def encode_transactions(data):
    te = TransactionEncoder()
    te_ary = te.fit(data).transform(data)
    return pd.DataFrame(te_ary, columns=te.columns_)
```

```python
# Apriori algorithm

from mlxtend.frequent_patterns import apriori, association_rules

def run_apriori_analysis(name, data):
    print(f"\n🔍 Running Apriori on {name} Dataset")

    df = encode_transactions(data)

    frequent_apriori = apriori(df, min_support=0.2, use_colnames=True)
    rules_apriori = association_rules(frequent_apriori, metric="confidence", min_threshold=0.6)

    print("\nApriori Frequent Itemsets:")
    print(frequent_apriori)

    print("\nApriori Association Rules:")
    print(rules_apriori[['antecedents', 'consequents', 'support', 'confidence', 'lift']])

    return frequent_apriori, rules_apriori

# Example calls (make sure space_data and sports_data are defined properly)
space_ap, space_ap_rules = run_apriori_analysis("Space", space_data)
sports_ap, sports_ap_rules = run_apriori_analysis("Sports", sports_data)
```

```python
# FP Algorithm
from mlxtend.frequent_patterns import fpgrowth, association_rules

def run_fpgrowth_analysis(name, data):
    print(f"\n Running FP-Growth on {name} Dataset")

    df = encode_transactions(data)

    frequent_fp = fpgrowth(df, min_support=0.2, use_colnames=True)
    rules_fp = association_rules(frequent_fp, metric="confidence", min_threshold=0.6)

    print("\n FP-Growth Frequent Itemsets:")
    print(frequent_fp)

    print("\n FP-Growth Association Rules:")
    print(rules_fp[['antecedents', 'consequents', 'support', 'confidence', 'lift']])

    return frequent_fp, rules_fp

space_fp, space_fp_rules = run_fpgrowth_analysis("Space", space_data)
sports_fp, sports_fp_rules = run_fpgrowth_analysis("Sports", sports_data)

```

```python

print("\n Comparison Summary:")

print(f"\nSpace Dataset:")
print(f"Apriori found {len(space_ap)} itemsets, {len(space_ap_rules)} rules.")
print(f"FP-Growth found {len(space_fp)} itemsets, {len(space_fp_rules)} rules.")

print(f"\nSports Dataset:")
print(f"Apriori found {len(sports_ap)} itemsets, {len(sports_ap_rules)} rules.")
print(f"FP-Growth found {len(sports_fp)} itemsets, {len(sports_fp_rules)} rules.")
```
